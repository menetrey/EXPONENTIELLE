# I) La fonction exponentielle
## 1) Définition et théorèmes

<div class="alert alert-block alert-danger">
 Il existe une unique fonction $f$ dérivable sur $\mathbb{R}$ telle que : $f′=f$ et $f(0) =1$. <br>
Cette fonction est appelée fonction exponentielle et on la note : $exp$.
</div>

<div class="alert alert-block alert-info">
L’existence de cette fonction est admise. Montrons que cette fonction ne s’annule pas sur $\mathbb{R}$.    
    
<b>La fonction exponentielle ne s’annule pas sur $\mathbb{R}$</b>
    
Soit la fonction $\phi$ définie sur $\mathbb{R}$ par :  
    $\phi(x) =f(x)f(−x)$.<br>  
Comme $f$ est dérivable sur $\mathbb{R}$, la fonction $\phi$ est dérivable sur $\mathbb{R}$ par produit :  <br> 
    $$\phi′(x) =f′(x)f(−x)−f(x)f′(−x) = f(x)f(−x)−f(x)f(−x) =0 $$ <br>
Comme $\phi′=0$ alors la fonction $\phi$ est constante. Donc : <br> 
    $$\forall x \in \mathbb{R}, \quad \phi(x) =\phi(0) =f^2(0) =1$$ <br>
 On en déduit alors : $f(x)f(−x) =1$, donc la fonction $f$ ne peut s’annuler.    
</div>

## 2) Approche graphique : méthode d’Euler
Déterminer un algorithme permettant de visualiser la fonctionexponentielle à partir de sa définition sur l’intervalle $[−a;a]$.   

On utilise l’approximation affine en $(x_0+p)$.   
$$\begin{array} 
 t f(x_0+p) \approx f(x_0) +p f′(x_0) &\iff& f(x_0+p)\approx f(x_0) +p f(x_0) \\
&\iff& f(x_0+p) \approx f(x_0)(1+p)
\end{array}$$  

L’approximation sera d’autant meilleure que $p$ sera petit.  
On commence à tracer le point $(0; 1)$ car $f(0)=1$, puis avec un pas $p$, on trace de proche en proche les points à droite $(X;Y)$ et les points à gauche $(−X;Z)$ dupoint $(0; 1)$ dans l’intervalle $[−a;a]$.

## III) Croissance et décroissance exponentielle
### 1) Modèle continu
<div class="alert alert-block alert-danger">
Soit un réel $k>0$, on définit les fonctions $f_k$ et $g_k$ sur $\mathbb{R}$ par :<br>
    $$f_k(t) =e^{kt}\quad \mbox{ et} \quad g_k(t) =e^{−k}t$$
Les fonctions $f_k$ correspondent à une croissance exponentielle.  
Les fonctions $g_k$ correspondent à une décroissance exponentielle ou atténuation.
</div>

Les fonctions $f_k$ et $g_k$ sont dérivables sur $\mathbb{R}$:  
$$\forall t \in \mathbb{R}, \quad f′_k(t) =ke^{kt} \gt 0 \quad \mbox{  et} \quad g′_k(x) =−ke^{−kt}\lt 0$$  
Les fonctions $f_k$ et $g_k$ sont respectivement croissantes et décroissantes.
